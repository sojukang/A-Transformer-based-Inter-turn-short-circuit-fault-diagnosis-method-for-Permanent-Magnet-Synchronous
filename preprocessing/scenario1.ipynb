## Make scenario

### healthy state

```python
healthy_merged_dataset = DivideAndMergeDataset.MergeSubsets(healthy_subset_4800[0])
healthy_merged_dataset.add_subset(healthy_subset_6000[0])
healthy_merged_dataset.add_subset(healthy_subset_7200[0])

healthy_merged_dataset.add_subset(healthy_subset_7200[1])
healthy_merged_dataset.add_subset(healthy_subset_6000[1])
healthy_merged_dataset.add_subset(healthy_subset_4800[1])

healthy_merged_dataset.add_subset(healthy_subset_4800[2])
healthy_merged_dataset.add_subset(healthy_subset_6000[2])
healthy_merged_dataset.add_subset(healthy_subset_7200[2])

healthy_merged_dataset.add_subset(healthy_subset_7200[3])
healthy_merged_dataset.add_subset(healthy_subset_6000[3])
healthy_merged_dataset.add_subset(healthy_subset_4800[3])
healthy_merged_dataset.dataset['fault'] = 0
```

### 1turn 1rf fault

```python
nsc_1turn_1rf_merged_dataset = DivideAndMergeDataset.MergeSubsets(nsc_1turn_1rf_subset_4800[0])
nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_6000[0])
nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_7200[0])

nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_7200[1])
nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_6000[1])
nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_4800[1])

nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_4800[2])
nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_6000[2])
nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_7200[2])

nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_7200[3])
nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_6000[3])
nsc_1turn_1rf_merged_dataset.add_subset(nsc_1turn_1rf_subset_4800[3])
nsc_1turn_1rf_merged_dataset.dataset['fault'] = 1
```

### 1turn 0.01rf fault

```python
nsc_1turn_merged_dataset = DivideAndMergeDataset.MergeSubsets(nsc_1turn_subset_4800[0])
nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_6000[0])
nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_7200[0])

nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_7200[1])
nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_6000[1])
nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_4800[1])

nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_4800[2])
nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_6000[2])
nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_7200[2])

nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_7200[3])
nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_6000[3])
nsc_1turn_merged_dataset.add_subset(nsc_1turn_subset_4800[3])
nsc_1turn_merged_dataset.dataset['fault'] = 1
```

### 5turns 1rf fault
